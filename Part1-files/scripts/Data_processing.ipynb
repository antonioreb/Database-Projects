## Tabela Albums <a name="albums"></a>


aqui tratamos a tabela Albums para ela ficar com os atributos correspondentes ao ER, para além de limparmos os dados

#### Construir o band_id

```python
import pandas as pd
```

importar a tabela bands

```python
bands  = pd.read_csv("../new_sets/Bands.csv", sep = ";")
bands
```

importar o csv dos albums (como não tem no header o nome das colunas temos que especificar)

```python
raw_albums = pd.read_csv("/home/antonio/Desktop/aulas/Bases_Dados/TBD-Project/raw_csv/band-album_data.csv", sep = ";", names = ["band_url","album_name", "release" , "abstract", "time","sales"],dtype= {"time":"object"})
```

```python
raw_albums.head()
```

mudar o formato da data do release de dd/mm/yyyy para dd-mm-yyyy (vai ser importante mais tarde)

```python
raw_albums["release"] = raw_albums["release"].apply(lambda x : x.replace("/","-"))
```

teste de duplicados

```python
raw_albums[raw_albums[["band_url","album_name","time","release","sales"]].duplicated(keep=False)]
```

não há duplicados, então podemos seguir em frente

Join das colunas que interessam do bands com o raw_albums para ficar com o band_id na tabela:

```python
albums = pd.merge(raw_albums,bands[["band_id","band_url"]]  )
```

```python
albums.head()
```

o band_url foi usado para fazer o join, mas já não é preciso, então removemos essa coluna

```python
albums.drop("band_url",axis = 1,inplace = True)
```

reorganizamos a ordem das colunas

```python
albums = albums.reindex(["band_id","album_name","release","sales","time" , "abstract"],axis = 1)
```

#### Limpeza da coluna "release"

explorando, descobrimos que há datas da coluna release que têm problemas como o dia do mês com o valor 0

```python
albums[(albums["release"] == "1967-12-0")]
```

```python
albums[albums.release.str.contains('-0$')]
```

```python
albums.at[27731,"release"] = "1967-12-1"
albums.at[29992,"release"] = "1982-09-4"
```

é importante transformar a release para o formato nativo de datetime no pandas porque facilita o processo de guardar no formato standard para o SQL de yyyy-mm-dd, então aqui fazemos isso usando o pd.to_datetime

```python
albums["release"] = pd.to_datetime(albums["release"],infer_datetime_format=True)
```

```python
albums.head()
```

teste de duplicados depois da limpeza

```python
albums[albums.duplicated(keep=False)]
```

#### Limpeza da coluna "sales"

existem algumas entradas defeituosas então temos que as detetar e limpar. No caso, aparecem umas quantas com "rowspan", como vemos aqui

```python
albums[albums.sales.str.contains("rowspan")]
```

os dados estão em formato string portanto temos que ser meticulosos para apenas deixar números e nada mais

```python
albums["sales"] = albums["sales"].apply(lambda x : x.replace( ",", "").replace(".","").replace( "rowspan=2 ","").replace( "rowspan=2","").replace( "rowspan=","").replace( "rowspan",""))
```

teste da limpeza

```python
albums[albums.sales.str.contains("rowspan")]
```

ainda assim, na última linha podem ver que temos "rowspan=250 000" como retiramos tudo o que fosse "rowspan=2" temos que repor o valor correto

```python
#restoring slowhand value
albums.at[29816,"sales"] = "250000"
```

calculo auxiliar para determinar o comprimento destas strings na base de dados

```python
#calculating max characters for album_name
max(albums.album_name.str.len())
```

#### Remoção de duplicados

ao converter a coluna de sales para inteiros, vamos verificar que há duplicados disfarçados por conta de espaços em branco

```python
#conversão
albums = albums.astype({"sales" : 'int32'})
```

agora removemos os duplicados na Key que nos interessa para manter a integridade da DB

```python
albums = albums[(albums[["band_id","album_name","release","sales","time"]].dupblicated() == False)]
```

testamos para ver como fica

```python
albums[(albums.duplicated()) ]
```

#### Gravação do csv

```python
albums.to_csv('../new_sets/Albums.csv', sep = ";", index = False)
```

## Tabela Belongs <a name="belongs"></a>
aqui tratamos a tabela belongs, inserindo as colunas necessárias e limpando os dados

#### Reformatação para remover as aspas

os dados estão desformatados porque têm o ";" no sítio errado e têm aspas que não permitem com que o pd.read_csv() faça uma leitura, então começamos por corrigir essa formatação

```python


with open("../raw_csv/band-member-member_name.csv", "r+") as file:
    lines = file.read()
    #print(lines.replace("\"",""))
    file.seek(0)
    #file.write(lines.replace(";\"","\";\"").replace("\"\"\"","\"\"").replace("\"\"","\""))
    file.write(lines.replace("\"",""))
    file.truncate()
      
```

```python
with open("../raw_csv/band-former_member-member_name.csv", "r+") as file:
    lines = file.read()
    #print(lines.replace("\"",""))
    file.seek(0)
    #file.write(lines.replace(";\"","\";\"").replace("\"\"\"","\"\"").replace("\"\"","\""))
    file.write(lines.replace("\"",""))
    file.truncate()
```

#### Importação dos ficheiros para os dataframes

"current" será o df com os artistas correntes, "former" o dos antigos artistas

```python
current = pd.read_csv("../raw_csv/band-member-member_name.csv", sep = ";", names = ["band_url", "Artist_URL","artist_name"])
former = pd.read_csv("../raw_csv/band-former_member-member_name.csv", sep = ";" ,names = ["band_url", "Artist_URL","artist_name"])
bands = pd.read_csv("../new_sets/Bands.csv", sep = ";")
artists = pd.read_csv("../new_sets/Artists.csv", sep = ";")
```

#### Formatação da coluna status
Começamos por adicionar uma coluna de True para o df current e uma coluna de False para o former

```python
truecolumn = pd.DataFrame({'Status':[True for x in range(current.shape[0])]}) #[_ for x in _] é uma list comprehension e current.shape[0] é numero de rows

current = pd.concat([current,truecolumn],axis=1) #aqui juntamos as duas, o axis indica que é uma coluna e não uma row
current
```

```python
falsecolumn = pd.DataFrame({'Status':[False for x in range(former.shape[0])]})

former = pd.concat([former,falsecolumn],axis=1)
former
```

agora juntamos current e former no mesmo df

```python
belongs = pd.concat([current,former])
```

#### Inserção do band_id
aqui fazemos um join (ou merge, não são muito diferentes) do Belongs com o df de Bands, para ter o band_id como coluna

```python
temp = pd.merge(belongs,bands[["band_url","band_id"]],left_on = "band_url",right_on = "band_url",how = "left")
#nota:os parametros opcionais dados são redundantes
```

```python
temp.shape
```

#### Inserção do Artist_id
aqui fazemos o merge para obter o Artist_id, mas agora com Artists

```python
artists.columns
```

```python
belongs = pd.merge(temp,artists)
```

```python
belongs.head()
```

```python
belongs.shape
```

as colunas com URL foram usadas para o merge, mas já não são mais precisas

```python
belongs.drop(["band_url","Artist_URL"],axis=1 , inplace=True) 
#o inplace faz as mudanças diretamente no df, se deixado como false a função .drop faria return de um novo df
```

reorganizamos a ordem das colunas

```python
belongs = belongs.reindex(columns = ["band_id", "Artist_ID", "artist_name" , "Status"])
```

### Limpeza dos dados

começamos por ordenar por artist_name para melhor ver os dados

```python
belongs.sort_values(by = ["artist_name"],inplace=True)
```

```python
belongs.head(50)
```

todas as entradas com "OBE" ou similares, são duplicadas, como descobrimos por exploração, então aqui serão removidas

```python
belongs = belongs[(belongs.artist_name != "(OBE)") & (belongs.artist_name != "(KBE)") & (belongs.artist_name != "(MBE)")]
```

agora limpamos os nomes que vêm entre parentesis ()

```python
belongs['artist_name'] = pd.concat(([belongs['artist_name'][:30].apply(lambda x : x.strip('()')),
                                                                    belongs['artist_name'][30:]]))
#não queremos tirar todos os parentesis dos nomes, apenas dos 30 primeiros
#Pode haver um nome como "António (Augusto) Rebelo" e nesse caso importa manter, por isso o concat
```

```python
belongs.head(30)
```

reordenamos e removemos os duplicados

```python
belongs.sort_values(by = ["artist_name"],inplace=True)
```

```python
belongs.drop_duplicates(inplace = True)
```

Será que há artistas que já saíram da banda e depois de ter ido para outra banda voltaram à original?

```python
belongs[belongs[["band_id","Artist_ID", "artist_name"]].duplicated(keep=False)]
```

Parece que sim. Como isto violaria a PK teremos que remover os duplicados com "status=False"

```python
belongs.drop((belongs[(belongs[["band_id","Artist_ID", "artist_name"]].duplicated(keep=False))
                      & (belongs["Status"] == False)].index, inplace = True))
```

#### Gravação dos dados

```python
belongs.to_csv("../new_sets/Belongs.csv",sep = ";" ,index=False)
```
